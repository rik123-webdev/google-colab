```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("/content/Salary_Data.csv")
print(df)
```

```python
sns.heatmap(df.isnull())
```

```python
x = df.iloc[:,:1].values
y = df.iloc[:,-1].values
```

```python
plt.scatter(x,y) #simple linear regression
```

```python
avgsal = df['Salary'].mean()
df.fillna({'Salary':avgsal},inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
LR_model = LinearRegression()
LR_model.fit(x,y)
```

```python
inp=[[3.5]]
y_pred = LR_model.predict(inp)
print(y_pred)
```

```python
plt.scatter(x,y)
plt.plot(x,LR_model.predict(x),color='red')
plt.title('Salary Graph')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.scatter(inp,y_pred,color='green')
plt.show()
```

```python
data = pd.read_csv('/content/Position_Salaries (1).csv')
print(data)
```

```python
sns.heatmap(data.isnull())
```

```python
data = data.drop(["Position"],axis=1)
print(data)
```

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0,shuffle=True)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
plt.scatter(x,y,color='green')
plt.plot(x,model.predict(x))
```

```python
from sklearn.preprocessing import PolynomialFeatures
mypoly = PolynomialFeatures(degree=4)
x_poly = mypoly.fit_transform(x)
print(x_poly)
```

```python
poly_model = LinearRegression()
poly_model.fit(x_poly,y)
```

```python
inp = [[2.9]]
poly_inp = mypoly.fit_transform(inp)
yp = poly_model.predict(poly_inp)
print(yp)
```

```python
plt.scatter(x,y,color='green')
plt.plot(x,poly_model.predict(x_poly))
plt.scatter(inp,yp,color='black')
```
