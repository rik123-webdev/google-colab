```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plane=pd.read_csv('airline_passengers.csv',index_col='Month',parse_dates=True)
plane.plot()
```

to calculate **trends** over years use
**SMA**
**EWMA**

```python
plane['12 month SMA']=plane['Thousands of Passengers'].rolling(window=12).mean()
plane['120 month SMA']=plane['Thousands of Passengers'].rolling(window=120).mean()
plane['1200 month SMA']=plane['Thousands of Passengers'].rolling(window=1200).mean()
plane['12000 month SMA']=plane['Thousands of Passengers'].rolling(window=12000).mean()
```

```python
plane
plane.plot()
```

```python
plane['EWMA12']=plane['Thousands of Passengers'].ewm(span=12,adjust=False).mean()
plane['EWMA120']=plane['Thousands of Passengers'].ewm(span=120,adjust=False).mean()
plane['EWMA1200']=plane['Thousands of Passengers'].ewm(span=1200,adjust=False).mean()
plane['EWMA12000']=plane['Thousands of Passengers'].ewm(span=12000,adjust=False).mean()
plane.plot()
```
